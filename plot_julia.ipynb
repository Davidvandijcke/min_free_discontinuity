```julia
using Plots
using JLD

data = load("./data.jld")
v = data["v"]
sigma = data["sigma"]
N = data["N"]
M = data["M"]

I = Array{Float64}(undef,N)
J = Array{Float64}(undef,N)
for i = 1:N
    I[i] = i/N
end
for k = 1:M
    J[k] = k/M
end

## 1D
# u = zeros(N)
# for i = 1:N
#     k = 1
#     while k < M && v[i,k] > 0.5
#         k = k + 1
#     end
#     u[i] = k/M
# end
# p = plot(I,u,xlim=(1/M,1),ylim=(1/M,1),zlim=(1/M,1),size=(500,300))   
# display(p)

## 2D
u = zeros(N,N);
for i = 1:N
   for j = 1:N
       k = 1
       while k < M && v[i,j,k] > 0.5
           k = k + 1
       end
       u[i,j] = k/M
   end
end
p = plot(I,I,u,seriestype=:surface,xlim=(1/M,1),ylim=(1/M,1),zlim=(1/M,1))
display(p)
```
