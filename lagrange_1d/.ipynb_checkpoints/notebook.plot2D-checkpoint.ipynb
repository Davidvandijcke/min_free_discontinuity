```julia
using Plots
gr()
```

```julia
using JLD
```

```julia
v = load("data_2d.jld")["v"];
N = load("data_2d.jld")["N"];
M = load("data_2d.jld")["M"];

I = zeros(N)
for i = 1:N
    I[i] = i/N
end
J = zeros(M)
for k = 1:M
    J[k] = k/M
end

u = zeros(N,N);
for i = 1:N
    for j = 1:N
        k = 1
        while k < M && v[i,j,k] > 0.5
            k = k + 1
        end
        u[i,j] = k/M
    end
end
```

```julia
plotly()
plot(I,I,u,seriestype=:surface,xlim=(1/M,1),ylim=(1/M,1),zlim=(2/M-1/100,1),size=(500,300))
```

```julia
u1 = zeros(N,N);
for i = 1:N
    for j = 1:N
        k = 1
        while k < M && v[i,j,k] > 0.99
            k = k + 1
        end
        u1[i,j] = k/M
    end
end
u2 = zeros(N,N);
for i = 1:N
    for j = 1:N
        k = 1
        while k < M && v[i,j,k] > 0
            k = k + 1
        end
        u2[i,j] = k/M
    end
end
```

```julia
plot(I,I,u1,seriestype=:surface,xlim=(1/M,1),ylim=(1/M,1),zlim=(2/M-0.001,1),size=(500,300))
plot!(I,I,u2,seriestype=:surface,xlim=(1/M,1),ylim=(1/M,1),zlim=(2/M-0.001,1),size=(500,300))
```

```julia
a = zeros(N,M)
for i = 1:N
    for k = 1:M
        a[i,k]=v[i,i,k]
    end
end
```

```julia
plot(I,J,a,seriestype=:surface,xlim=(1/M,1),ylim=(1/M,1),zlim=(0,1),size=(500,300))
```
