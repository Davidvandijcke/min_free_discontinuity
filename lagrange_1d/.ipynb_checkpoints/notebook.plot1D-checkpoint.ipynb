```julia
using Plots
gr()
```

```julia
using JLD
```

```julia
v = load("data_1d.jld")["v"];
N = load("data_1d.jld")["N"];
M = load("data_1d.jld")["M"];

I = zeros(N)
for i = 1:N
    I[i] = i/N
end
J = zeros(M)
for k = 1:M
    J[k] = k/M
end

u = zeros(N);
for i = 1:N
    k = 1
    while k < M && v[i,k] > 0.5
        k = k + 1
    end
    u[i] = k/M
end
plot(I,u,xlim=(1/M,1),ylim=(1/M,1),zlim=(2/M,1),size=(500,300))
```

```julia
u1 = zeros(N);
for i = 1:N
        k = 1
        while k < M && v[i,k] > 0.99
            k = k + 1
        end
        u1[i] = k/M
end
u2 = zeros(N);
for i = 1:N
    k = 1
    while k < M && v[i,k] > 0
        k = k + 1
    end
    u2[i] = k/M
end
plot(I,u1,xlim=(1/M,1),ylim=(1/M,1),zlim=(2/M,1),size=(500,300))
plot!(I,u2,xlim=(1/M,1),ylim=(1/M,1),zlim=(2/M,1),size=(500,300))
```

```julia
plot(I,J,v,seriestype=:surface,xlim=(1/M,1),ylim=(1/M,1),zlim=(0,1),size=(500,300))
```
